0)	Código contendo as funções definidas para resolver todos os exercícios abaixo. É necessário executar a célula abaixo antes de qualquer outra.

```python
import sys
import cv2
import matplotlib.pyplot as plt
import numpy as np

def convolution(image, kernel):
    image_row, image_col = image.shape
    kernel_row, kernel_col = kernel.shape

    output = np.zeros(image.shape)

    pad_height = 1
    pad_width = 1

    padded_image = np.zeros((image_row + (2 * pad_height), image_col + (2 * pad_width)))
    padded_image[pad_height:padded_image.shape[0] - pad_height, pad_width:padded_image.shape[1] - pad_width] = image

    for row in range(image_row):
        for col in range(image_col):
            output[row, col] = np.sum(kernel * padded_image[row:row + kernel_row, col:col + kernel_col])
    
    return output
```

1) Aplique os filtros de realce Sobel e Roberts nas imagens abaixo. Mostre, para o caso do Sobel, uma imagem representando a derivada em x e outra imagem representando a derivada em y

```python
np.set_printoptions(threshold=sys.maxsize)

# le imagem
img = cv2.imread("cln1.png", cv2.IMREAD_GRAYSCALE)

# aplica filtro de Sobel
sobel_filter_x = np.array(
    [[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
sobel_image_x = convolution(img, sobel_filter_x)
sobel_filter_y = np.array(
    [[1, 2, 1], [0, 0, 0], [-1, -2, -1]])
sobel_image_y = convolution(img, sobel_filter_y)
sobel = np.sqrt(sobel_image_x ** 2 + sobel_image_y ** 2)

# aplica filtro de roberts
kernelx = np.array([[1, 0], [0, -1]])
kernely = np.array([[0, 1], [-1, 0]])
roberts_image_x = convolution(img, kernelx)
roberts_image_y = convolution(img, kernely)
roberts = np.sqrt(roberts_image_x ** 2 + roberts_image_y ** 2)

fig, axs = plt.subplots(2, 4, figsize=(10, 7))
axs[0][0].imshow(img, cmap='gray')
axs[0][0].set_title('Original Image')
axs[0][1].imshow(sobel_image_x, cmap='gray')
axs[0][1].set_title('Sobel X')
axs[0][2].imshow(sobel_image_y, cmap='gray')
axs[0][2].set_title('Sobel Y')
axs[0][3].imshow(sobel, cmap='gray')
axs[0][3].set_title('Sobel')
axs[1][0].imshow(img, cmap='gray')
axs[1][0].set_title('Original Image')
axs[1][1].imshow(roberts_image_x, cmap='gray')
axs[1][1].set_title('Roberts X')
axs[1][2].imshow(roberts_image_y, cmap='gray')
axs[1][2].set_title('Roberts Y')
axs[1][3].imshow(roberts, cmap='gray')
axs[1][3].set_title('Roberts')
plt.show()
```

2) Utilizando a imagem do exercício anterior, realce as bordas utilizando (a) Filtro Laplaciano. (b) Unsharp masking. (c) Highboost filtering (escolher um k). (d) K <1 para atenuar

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# le imagem
img = cv2.imread("cln1.png", cv2.IMREAD_GRAYSCALE)

# a - filtro laplaciano
laplacian_interior = convolution(img, np.array([[0, -1, 0], [-1, 4, -1], [0, -1, 0]]))
laplacian_exterior = convolution(img, np.array([[0, 1, 0], [1, -4, 1], [0, 1, 0]]))

# b - unsharp masking
k = 1
gmask = img - 0.5*laplacian_exterior
unsharp = img + k * gmask

# # c - highboost filtering
k2 = 2
highboost = img + k2 * gmask

# d - attenuation
k3 = 0.1
attenuation = img + k3 * gmask

fig, axs = plt.subplots(2, 3, figsize=(10, 7))
axs[0][0].imshow(img, cmap='gray')
axs[0][0].set_title('Original Image')
axs[0][1].imshow(laplacian_interior, cmap='gray')
axs[0][1].set_title('Laplacian Interior 4')
axs[0][2].imshow(laplacian_exterior, cmap='gray')
axs[0][2].set_title('Laplacian Exterior -4')
axs[1][0].imshow(unsharp, cmap='gray')
axs[1][0].set_title('Unsharp Masking')
axs[1][1].imshow(highboost, cmap='gray')
axs[1][1].set_title('Highboost')
axs[1][2].imshow(attenuation, cmap='gray')
axs[1][2].set_title('Attenuation')
plt.show()
```

3) Aplique o operador Sobel imagem original abaixo e obtenha as outras imagens conforme descrição. We can see that the intensity of many pixels on the surface is as high as along the actual edges. One reason is that the output of many edge pixels is greater than the maximum pixel value and therefore they are `cut off' at 255. To avoid this overflow we scale the range image by a factor 0.25 prior to the edge detection and then normalize the output, as can be seen in (c)

```python
# le imagem
img = cv2.imread("ren1.png", cv2.IMREAD_GRAYSCALE)

# aplica scale na imagem
scaled_img = cv2.convertScaleAbs(img, alpha=0.25)

# filtro de sobel
sobel_filter_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
sobel_image_x = convolution(scaled_img, sobel_filter_x)
sobel_filter_y = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])
sobel_image_y = convolution(scaled_img, sobel_filter_y)
sobel = np.sqrt(sobel_image_x ** 2 + sobel_image_y ** 2)

fig, axs = plt.subplots(2, 3, figsize=(10, 7))
axs[0][0].imshow(img, cmap='gray')
axs[0][0].set_title('Original Image')
axs[0][1].imshow(scaled_img, cmap='gray')
axs[0][1].set_title('Scaled Image')
axs[0][2].imshow(sobel_image_x, cmap='gray')
axs[0][2].set_title('Sobel X')
axs[1][0].imshow(sobel_image_y, cmap='gray')
axs[1][0].set_title('Sobel Y')
axs[1][1].imshow(sobel, cmap='gray')
axs[1][1].set_title('Sobel')
```

4) Aplique o operador Sobel imagem original abaixo e obtenha a imagem binária após a limiarização utilizando o valor 150

```python
# le imagem
img = cv2.imread('ufo2.png', cv2.IMREAD_GRAYSCALE)

# filtro de sobel
sobel_filter_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
sobel_image_x = convolution(img, sobel_filter_x)
sobel_filter_y = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])
sobel_image_y = convolution(img, sobel_filter_y)
sobel = np.sqrt(sobel_image_x ** 2 + sobel_image_y ** 2)

# binariza a imagem
maxval = 255
threshold = 150
binary = (img > threshold) * maxval

# resultados
fig, axs = plt.subplots(2, 3, figsize=(7, 7))
axs[0][0].imshow(img, cmap='gray')
axs[0][0].set_title('Original Image')
axs[0][1].imshow(sobel_image_x, cmap='gray')
axs[0][1].set_title('Sobel X')
axs[0][2].imshow(binary, cmap='gray')
axs[0][2].set_title('Binary')
axs[1][0].imshow(sobel_image_y, cmap='gray')
axs[1][0].set_title('Sobel Y')
axs[1][1].imshow(sobel, cmap='gray')
axs[1][1].set_title('Sobel')
plt.show()
```
