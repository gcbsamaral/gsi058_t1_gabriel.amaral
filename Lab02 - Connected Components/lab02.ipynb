```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

def image_grayscale(image):
    
    return np.where(image > 2, image*(256/(np.max(image)-2)), image-1)

def pile(image, line, column, label, stack):
    if image[line,column] == 1:
        coordinates = [line,column]
        stack.append(coordinates)
        image[line,column] = label
        
        return True
    
    return False

def connected_components(image):
    
    #Cria uma imagem resultante que começa como uma cópia da imagem original
    result_image = np.copy(image)
    #Cria uma pilha
    stack = []
    #Cria um rótulo que começa em 2
    label = 2
    
    #Itera pelos pixels da imagem já segmentada, agrupa os pixels em componentes conexos e aplica um rótulo para cada componente.
    for line_it in range(result_image.shape[0]):
        for column_it in range(result_image.shape[1]):
            if pile(result_image, line_it, column_it, label, stack):
                while stack: 
                    coordinates = stack.pop() #retira o elemento do topo da pilha e salva suas coordenadas
                    pile(result_image, coordinates[0], coordinates[1]+1, label, stack) #chama a função pile para o pixel da esquerda
                    pile(result_image, coordinates[0], coordinates[1]-1, label, stack) #chama a função pile para o pixel da direita
                    pile(result_image, coordinates[0]-1, coordinates[1], label, stack) #chama a função pile para o pixel acima
                    pile(result_image, coordinates[0]+1, coordinates[1], label, stack) #chama a função pile para o pixel abaixo

                label+=1
    
    return result_image
```

```python
def image_segment(image, threshold):
    
    return np.where(image < threshold, 0, 1)

#Lê a imagem original
imagem = cv2.imread("art8.png", cv2.IMREAD_GRAYSCALE)
plt.plot(3, 1, 1)
plt.imshow(imagem, cmap='gray', vmin=0, vmax=255)
plt.title("Imagem Original")
plt.show()

#Gera uma imagem rotulada com base em seus componentes conexos pré ajustada para que os componentes apareçam em escala de cinza
imagem_resultado = image_grayscale(connected_components(image_segment(imagem, 255)))

#Mostrando a imagem resultado em grayscale
plt.plot(3, 1, 2)
plt.imshow(imagem_resultado, cmap='gray', vmin=0, vmax=255)
plt.title("Imagem Resultado - Grayscale")
plt.show()

#Mostra a imagem resultado em cores
plt.plot(3, 1, 3)
plt.imshow(imagem_resultado, cmap='jet', vmin=0, vmax=255)
plt.title("Imagem Resultado - Colorida")
plt.show()
```
